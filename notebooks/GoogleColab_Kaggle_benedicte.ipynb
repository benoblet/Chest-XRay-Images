````
meta_df["patientID"] = ""
meta_df["patimgNUM"] = 0
for i, pictname in tqdm.tqdm(enumerate(meta_df["name"]),
                             unit='noms', total=5856):
  category = meta_df.loc[i, "class_str"]
  if category == "PNEUMONIA":
    infos = pictname.split("_")
    # c'est presque simple, il y a peu de double image
    nb_zeros = 10 - len(infos[0])
    case = "p"
    meta_df.loc[i, "patientID"] = case + "0"*nb_zeros + infos[0][6:]
    if len(infos) == 3:
      meta_df.loc[i, "patimgNUM"] = 1
    else:
      meta_df.loc[i, "patimgNUM"] = 1 + int(infos[-1])
  else:
    infos = pictname.split("-")
    # les normaux sont moins nombreux mais plus compliqués
    expected = 4
    case = "N"
    if pictname.startswith("IM"):
      expected = 3
      case ="n"
    if len(infos) == expected:
      meta_df.loc[i, "patientID"] = case + infos[-2]
      meta_df.loc[i, "patimgNUM"] = int(infos[-1])
    else:
      meta_df.loc[i, "patientID"] = case + infos[-3]
      meta_df.loc[i, "patimgNUM"] = int(infos[-1]) + int(infos[-2])
````

````
from keras import backend as K
from sklearn.metrics import roc_curve, precision_recall_curve, auc, confusion_matrix, ConfusionMatrixDisplay

import matplotlib.pyplot as plt
````

````
# Aller chercher la référence

def recall_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + K.epsilon())
    return recall

def precision_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision

def f1_m(y_true, y_pred):
    precision = precision_m(y_true, y_pred)
    recall = recall_m(y_true, y_pred)
    return 2 * ((precision * recall) / (precision + recall + K.epsilon()))
````

````
def plot_cm(cm):
    labels = ['NORMAL', 'PNEUMONIA']
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    disp.plot(cmap=plt.cm.Blues)
    plt.show()
````

````
def plot_roc_pr(y_test, y_pred):
    fig, ax = plt.subplots(1, 2, figsize=(12, 4))   
    
    fp_rate, tp_rate, thresholds_roc = roc_curve(y_test, y_pred)
    auc_keras = auc(fp_rate, tp_rate)
    ax[0].plot([0, 1], [0, 1], linestyle='--')
    ax[0].plot(fp_rate, tp_rate, marker='.')
    ax[0].set_title('Courbe ROC')
    ax[0].set_xlabel('Taux de faux positifs')
    ax[0].set_ylabel('Taux de vrais positifs')
    ax[0].legend(['Référence\nAUC : 0.5', f'Modèle\nAUC : {auc_keras:.3f}'], loc='lower right')
    
    precision, recall, thresholds = precision_recall_curve(y_test, y_pred)
    # TODO : faire calcul AUC PR
    ax[1].plot([1, 0], [0, 1], linestyle='--')
    ax[1].plot(recall, precision)
    ax[1].set_title('Courbe PR')
    ax[1].set_xlabel('Recall')
    ax[1].set_ylabel('Precision')
    ax[1].legend(['Référence\nAUC : 0.5', f'Modèle\nAUC : {auc_keras:.3f}'], loc='lower left')    
    
    plt.show()
    return auc_keras
````

````
def plot_loss_acc(history):
    fig, ax = plt.subplots(1, 2, figsize=(12, 4))
    
    ax[0].plot(history.history['loss'])
    ax[0].plot(history.history['val_loss'])
    ax[0].set_title('Loss du modèle')
    ax[0].set_xlabel('Époque')
    ax[0].set_ylabel('Loss')
    ax[0].legend(['train', 'val'], loc='upper right')

    ax[1].plot(history.history['accuracy'])
    ax[1].plot(history.history['val_accuracy'])
    ax[1].set_title('Accuracy du modèle')
    ax[1].set_xlabel('Époque')
    ax[1].set_ylabel('Accuracy')
    ax[1].legend(['train', 'val'], loc='lower right')
    
    fig.tight_layout()
    plt.show()
````

````
from sklearn import metrics
#roc_curve, precision_recall_curve, auc, confusion_matrix, ConfusionMatrixDisplay

import matplotlib.pyplot as plt

def plot_cm(cm):
    labels = ['NORMAL', 'PNEUMONIA']
    disp = metrics.ConfusionMatrixDisplay(confusion_matrix=cm,
                                          display_labels=labels)
    disp.plot(cmap=plt.cm.Blues)
    plt.show()

def plot_roc_pr(y_test, y_pred):
    fig, ax = plt.subplots(1, 2, figsize=(12, 4))   
    
    fp_rate, tp_rate, thresholds_roc = metrics.roc_curve(y_test, y_pred)
    auc_keras = metrics.auc(fp_rate, tp_rate)
    ax[0].plot([0, 1], [0, 1], linestyle='--')
    ax[0].plot(fp_rate, tp_rate, marker='.')
    ax[0].set_title('Courbe ROC')
    ax[0].set_xlabel('Taux de faux positifs')
    ax[0].set_ylabel('Taux de vrais positifs')
    ax[0].legend(['Référence\nAUC : 0.5', f'Modèle\nAUC : {auc_keras:.3f}'], loc='lower right')
    
    precision, recall, thresholds = metrics.precision_recall_curve(y_test, y_pred)
    # TODO : faire calcul AUC PR
    ax[1].plot([1, 0], [0, 1], linestyle='--')
    ax[1].plot(recall, precision)
    ax[1].set_title('Courbe PR')
    ax[1].set_xlabel('Recall')
    ax[1].set_ylabel('Precision')
    ax[1].legend(['Référence\nAUC : 0.5', f'Modèle\nAUC : {auc_keras:.3f}'], loc='lower left')    
    
    plt.show()
    return auc_keras

def plot_loss_acc(history):
    fig, ax = plt.subplots(1, 2, figsize=(12, 4))
    
    ax[0].plot(history.history['loss'])
    ax[0].plot(history.history['val_loss'])
    ax[0].set_title('Loss du modèle')
    ax[0].set_xlabel('Époque')
    ax[0].set_ylabel('Loss')
    ax[0].legend(['train', 'val'], loc='upper right')

    ax[1].plot(history.history['accuracy'])
    ax[1].plot(history.history['val_accuracy'])
    ax[1].set_title('Accuracy du modèle')
    ax[1].set_xlabel('Époque')
    ax[1].set_ylabel('Accuracy')
    ax[1].legend(['train', 'val'], loc='lower right')
    
    fig.tight_layout()
    plt.show()
````
