### 3. Fichiers fournis
Comptage des effectifs de fichiers par sous-dossier :

```bash
ls $rawdata_folder
```

```bash
logfile="${project_home}logfiles/data_discovery.log"

for sampling in $(ls ${rawdata_folder}); do
    folder="${rawdata_folder}${sampling}/"
    for group in $(ls "${folder}"); do
        subfolder="${folder}${group}/"
        echo "${subfolder}" | tee -a ${logfile}
        echo "$(ls ${subfolder} | wc -l) lines" | tee -a ${logfile}
    done
done
```

Noms de fichiers :

```bash
logfile="${project_home}logfiles/data_discovery.log"

echo "=== Noms des fichiers ===" | tee -a ${logfile}

for sampling in $(ls ${rawdata_folder}); do
    folder="${rawdata_folder}${sampling}/"
    
    for group in $(ls "${folder}"); do
        subfolder="${folder}${group}/"
        echo "${group}" | tee -a ${logfile}
        echo "$(ls ${subfolder})" | tee -a ${logfile}
        
    done
done
```
